```python
import openai
import os
```

```python
def generate_music_info(prompt):
    response = openai.ChatCompletion.create(
        model='gpt-3.5-turbo',
        messages=[
            {'role': 'user', 'content': prompt}
        ],
        max_tokens=150
    )

    return response['choices'][0]['message']['content']

def main():
    # 사용자 입력 받기
    user_input = input('음원에 대한 키워드를 입력하세요: ')

    # 프롬프트 구성
    prompt = (
        f'\'{user_input}\'라는 키워드를 바탕으로 음악을 생성하기 위한 상세 정보를 다음과 같은 형식으로 구성해 주세요:\n\n'
        f'1. **장르**: 생성할 음악의 장르를 명확하게 정의해주세요. 예를 들어, \'팝\', \'록\', \'재즈\', \'클래식\' 등으로 구체화해주세요.\n'
        f'   - **예시**: \'일렉트로닉 팝\'\n\n'
        f'2. **리듬**: 음악의 리듬을 설명해주세요. 예를 들어, \'빠른\', \'느린\', \'미디엄 템포\', \'스윙\' 등으로 정의해주세요.\n'
        f'   - **예시**: \'120 BPM의 미디엄 템포\'\n\n'
        f'3. **멜로디**: 생성할 음악의 멜로디에 대한 설명을 추가해주세요. 예를 들어, \'경쾌하고 부드러운 멜로디\', \'강렬하고 에너제틱한 멜로디\' 등으로 표현해주세요.\n'
        f'   - **예시**: \'감정을 자극하는 상승하는 멜로디\'\n\n'
        f'4. **테마**: 음악의 주제나 감정을 설명해주세요. 예를 들어, \'사랑\', \'이별\', \'축제\', \'자연\', \'모험\' 등으로 구체화해주세요.\n'
        f'   - **예시**: \'여름의 자유로운 순간들\'\n\n'
        f'5. **악기**: 사용하고자 하는 악기 목록을 포함해주세요. 예를 들어, \'기타\', \'피아노\', \'드럼\', \'신디사이저\', \'바이올린\' 등으로 명시해주세요.\n'
        f'   - **예시**: \'어쿠스틱 기타, 드럼 머신, 신디사이저\'\n\n'
        f'6. **참조 곡**: 비슷한 스타일의 기존 곡을 제시해주세요. 예를 들어, \'Dua Lipa - Levitating\', \'Kygo - Firestone\'과 같은 곡을 언급해주세요.\n'
        f'   - **예시**: \'Ed Sheeran - Shape of You\'\n\n'
        f'**추가 지시사항**:\n'
        f'1. 각 항목을 구체적으로 작성하여 음악의 방향성을 명확히 해주세요.\n'
        f'2. 가사의 내용과 분위기를 음악에 잘 반영하도록 하세요.\n'
        f'3. 지정된 장르의 특성과 스타일을 살려주세요.\n'
        f'4. 음악의 구조나 형식(예: 인트로, 벌스, 코러스, 아웃트로 등)에 대한 설명도 포함해주세요. \n'
        f'   - **예시**: \'인트로 (4마디) → 벌스 (8마디) → 코러스 (8마디) → 아웃트로 (4마디)\'\n'
    )

    # 음원 정보 생성
    music_info = generate_music_info(prompt)

    # 반환된 내용 출력
    print('API로부터 받은 응답:')
    print(music_info)

    # 생성된 음원 상세 정보를 문자열로 변환
    music_info_lines = music_info.split('\n')

    # 각 항목에 대한 안전한 접근
    genre = music_info_lines[0].split(': ')[1] if len(music_info_lines) > 0 and ': ' in music_info_lines[0] else '알 수 없음'
    rhythm = music_info_lines[1].split(': ')[1] if len(music_info_lines) > 1 and ': ' in music_info_lines[1] else '알 수 없음'
    melody = music_info_lines[2].split(': ')[1] if len(music_info_lines) > 2 and ': ' in music_info_lines[2] else '알 수 없음'
    theme = music_info_lines[3].split(': ')[1] if len(music_info_lines) > 3 and ': ' in music_info_lines[3] else '알 수 없음'
    instruments = music_info_lines[4].split(': ')[1] if len(music_info_lines) > 4 and ': ' in music_info_lines[4] else '알 수 없음'
    reference_song = music_info_lines[5].split(': ')[1] if len(music_info_lines) > 5 and ': ' in music_info_lines[5] else '알 수 없음'

    music_info_str = (
        f'장르: {genre}\n'
        f'리듬: {rhythm}\n'
        f'멜로디: {melody}\n'
        f'테마: {theme}\n'
        f'악기: {instruments}\n'
        f'참조 곡: {reference_song}\n'
    )

    print('생성된 음원 상세 정보:')
    print(music_info_str)

# 실행
if __name__ == '__main__':
    main()
```
